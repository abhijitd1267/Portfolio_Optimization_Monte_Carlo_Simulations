```python
# Function to plot financial data using Plotly express
def plot_financial_data(df, title):
    
    fig = px.line(title = title)
    
    # Starting index is 1 to skip the 'Date' column
    for i in df.columns[1:]:
        fig.add_scatter(x = df['Date'], y = df[i], name = i)
        fig.update_traces(line_width = 5)
        fig.update_layout({'plot_bgcolor': 'white'})
        
    fig.show()              
```

```python
plot_financial_data(stock_df.drop(columns=['Volume', 'Daily Return']), 'Amazon.com, Inc. (AMZN) Stock Price [$]')
```

```python
# Plotting volume data
plot_financial_data(stock_df.iloc[:, [0, 5]], 'Amazon.com, Inc. (AMZN) Trading Volume')
```

```python
# Plotting daily return data
plot_financial_data(stock_df.iloc[:, [0, 7]], 'Amazon.com, Inc. (AMZN) Daily Return [%]')
```

# Data visualization of a single stock: Part 2

```python
# Function to classify percentage daily return of a stock

def classify_daily_return(percentage_return):
    if percentage_return > -0.3 and percentage_return <= 0.3:
        return 'Insignificant Change'
    elif percentage_return > 0.3 and percentage_return <= 3:
        return 'Positive Change'
    elif percentage_return > -3 and percentage_return <= -0.3:
        return 'Negative Change'
    elif percentage_return > 3 and percentage_return <= 7:
        return 'Large Positive Change'
    elif percentage_return > -7 and percentage_return <= -3:
        return 'Large Negative Change'
    elif percentage_return > 7:
        return 'Bull Run'
    elif percentage_return <= -7:
        return 'Bear Sell Off'
```

```python
stock_df['Trend'] = stock_df['Daily Return'].apply(classify_daily_return)
stock_df
```

```python
trend_summary = stock_df['Trend'].value_counts()
trend_summary
```

```python
# Pie chart of trend summary
plt.figure(figsize=(8, 8))
trend_summary.plot(kind = 'pie', y = 'Trend', title='Amazon.com, Inc. (AMZN) Daily Return Trend')
```

```python
JPM_df['Trend'] = JPM_df['Daily Return'].apply(classify_daily_return)
JPM_df
```

```python
trend_summary_JPM = JPM_df['Trend'].value_counts()
trend_summary_JPM
```

```python
plt.figure(figsize=(8, 8))
trend_summary_JPM.plot(kind = 'pie', y = 'Trend', title='JPMorgan Chase & Co. (JPMC) Daily Return Trend')
plt.show()
```
